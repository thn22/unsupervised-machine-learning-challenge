```python
import pandas as pd
import matplotlib.pyplot as plt
from pathlib import Path
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
from sklearn.manifold import TSNE
from sklearn.cluster import KMeans
```

```python
file_path = Path("Resources/myopia.csv")
df = pd.read_csv(file_path)
df
```

```python
labels = df["MYOPIC"]
```

```python
df = df.drop(columns=["MYOPIC"])
df.head()
```

```python
scaler = StandardScaler()
scaled_data = scaler.fit_transform(df)
```

```python
pca = PCA(n_components=0.9)

df_pca = pca.fit_transform(df)

pca.explained_variance_ratio_
```

```python
tsne = TSNE(learning_rate=250)
tsne_features = tsne.fit_transform(df_pca)
```

```python
plt.scatter(tsne_features[:,0], tsne_features[:,1], c=labels)
plt.show()
```

There are not any distinct clusters within this scatter plot.

```python
inertia = []
k = list(range(1,11))

for i in k:
    km = KMeans(n_clusters=i, random_state=0)
    km.fit(df_pca)
    inertia.append(km.inertia_)
    
elbow_data = {"k": k, "inertia": inertia}
df_elbow = pd.DataFrame(elbow_data)
df_elbow.head()
```

```python
plt.plot(df_elbow["k"], df_elbow["inertia"])
plt.xticks(range(1,11))
plt.xlabel("Number of Clusters")
plt.ylabel("Inertia")
plt.title("Elbow Curve for Myopia")
plt.show()
```

The elbow of the plot is at 83541.57 inertia and the 3rd value for k.

```python
columns = []
for i in range(len(pca.components_)):
    columns.append(i+1)
    
    
myopia_df = pd.DataFrame(df_pca, columns=columns)
myopia_df["class"] = km.labels_
myopia_df.head()
```

```python
model = KMeans(n_clusters=3, random_state=0)
model.fit(myopia_df)
predictions = model.predict(myopia_df)
myopia_df["class"] = model.labels_

plt.scatter(x=myopia_df[1], y=myopia_df[2], c=myopia_df["class"])
plt.xlabel("PCA 1")
plt.ylabel("PCA 2")
plt.title("Myopia Clusters")
plt.show()
```

We can clearly identify 3 distinct clusters that divide the patients.
